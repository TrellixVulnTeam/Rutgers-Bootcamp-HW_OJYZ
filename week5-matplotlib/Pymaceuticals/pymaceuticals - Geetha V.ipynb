```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data_to_load = "data/mouse_drug_data.csv"
drug_data_to_load = "data/clinicaltrial_data.csv"

mouse_data = pd.read_csv(mouse_data_to_load)
drug_data = pd.read_csv(drug_data_to_load)
# Combine the data into a single dataset
trial_data_complete = pd.merge(drug_data, mouse_data, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
trial_data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
#gf3 = purchase_frame.groupby(['Gender','SN']).sum()["Price"] 
t1 = trial_data_complete.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"].reset_index(name="Tumor Vol")

# Convert to DataFrame

# Preview DataFrame
t1.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sef = trial_data_complete.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"].reset_index(name="Tumor Vol SE")
sef.head()
# Convert to DataFrame

# Preview DataFrame
#t2 = pd.concat([t1,sef],ignore_index=True)
#t2.head()

sef2 = sef.pivot(index="Timepoint", columns="Drug", values="Tumor Vol SE")
sef2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#t2 = t1[["Item ID", "Item Name"]]
t2 = t1.pivot(index="Timepoint", columns="Drug", values="Tumor Vol")
t2.head()
# Preview that Reformatting worked

```

```python
#create lists from rows and columns

# Generate the Plot (with Error Bars)
tp = t1["Timepoint"].unique()
capy = t2["Capomulin"].values
secapy = sef2["Capomulin"].values
infy = t2["Infubinol"].values
seinfy = sef2["Infubinol"].values
kety = t2["Ketapril"].values
sekety = sef2["Ketapril"].values
play = t2["Placebo"].values
seplay = sef2["Placebo"].values

plt.errorbar(tp, capy,xerr= secapy, marker='o',color='red', label="Capomulin")
plt.errorbar(tp, infy,xerr= seinfy,marker='^',color='blue', label="Infubinol")
plt.errorbar(tp, kety,xerr= sekety,marker='s',color='green', label="Ketapril")
plt.errorbar(tp, play,xerr= seplay,marker='d',color='black', label="Placebo")


plt.legend()
plt.title("Tumor Response to Treatment")
plt.xlabel("Total Time (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(True)

# Save the Figure
from io import BytesIO
f = "tumor.jpg"
plt.savefig(f, format="jpg")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
m1 = trial_data_complete.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"].reset_index(name="Avg Meta Sites")
m1.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
semeta = trial_data_complete.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"].reset_index(name="Metastatic Sites SE")
semeta.head()
# Convert to DataFrame
semeta2 = semeta.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites SE")
semeta2.head()
# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
m2 = m1.pivot(index="Timepoint", columns="Drug", values="Avg Meta Sites")
m2.head()
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)
tp = t1["Timepoint"].unique()
capy = m2["Capomulin"].values
secapy = semeta2["Capomulin"].values
infy = m2["Infubinol"].values
seinfy = semeta2["Infubinol"].values
kety = m2["Ketapril"].values
sekety = semeta2["Ketapril"].values
play = m2["Placebo"].values
seplay = semeta2["Placebo"].values

plt.errorbar(tp, capy,yerr= secapy, marker='o',color='red', label="Capomulin")
plt.errorbar(tp, infy,yerr= seinfy,marker='^',color='blue', label="Infubinol")
plt.errorbar(tp, kety,yerr= sekety,marker='s',color='green', label="Ketapril")
plt.errorbar(tp, play,yerr= seplay,marker='d',color='black', label="Placebo")


plt.legend()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Total Time (days)")
plt.ylabel("Met. Sites")
plt.grid(True)
# Save the Figure

from io import BytesIO
f = "meta.jpg"
plt.savefig(f, format="jpg")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse1 = trial_data_complete.groupby(["Drug", "Timepoint"]).nunique()["Mouse ID"].reset_index(name="Mouse Count")
mouse1.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse2 = mouse1.pivot(index="Timepoint", columns="Drug", values="Mouse Count")
mouse2.head()
# Preview the Data Frame
```

```python
drugs = trial_data_complete['Drug'].unique().tolist()
drugs.sort()
drug_length = len(drugs)
cappct = []
infpct = []
ketpct = []
plapct = []
temp = []

test = mouse2.loc[[0]]
test
tp = t1["Timepoint"].unique()
tpl = len(tp)
tpl

for i in range(tpl): 
    temp = mouse2["Capomulin"].values
    dayone = mouse2.iloc[0]["Capomulin"]
    v = ((temp[i])/dayone) * 100
    cappct.append(v)
    

for i in range(tpl): 
    temp = mouse2["Infubinol"].values
    dayone = mouse2.iloc[0]["Infubinol"]
    v = ((temp[i])/dayone) * 100
    infpct.append(v)

for i in range(tpl): 
    temp = mouse2["Ketapril"].values
    dayone = mouse2.iloc[0]["Ketapril"]
    v = ((temp[i])/dayone) * 100
    ketpct.append(v)

for i in range(tpl): 
    temp = mouse2["Placebo"].values
    dayone = mouse2.iloc[0]["Placebo"]
    v = ((temp[i])/dayone) * 100
    plapct.append(v)

```

```python
# Generate the Plot (Accounting for percentages)
tp = t1["Timepoint"].unique()

plt.plot(tp, cappct,marker='o',color='red', label="Capomulin")
plt.plot(tp, infpct,marker='^',color='blue', label="Infubinol")
plt.plot(tp, ketpct,marker='s',color='green', label="Ketapril")
plt.plot(tp, plapct,marker='d',color='black', label="Placebo")


plt.legend()
plt.title("Survival During Treatment")
plt.xlabel("Total Time (days)")
plt.ylabel("Survival Rate (%)")
plt.grid(True)
# Save the Figure
from io import BytesIO
f = "survival.jpg"
plt.savefig(f, format="jpg")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
drugs = trial_data_complete['Drug'].unique().tolist()
drugs.sort()
drug_length = len(drugs)
rowzerolist = []
rowonelist = []
pctlist = []

test = t2.loc[[0, 45]]


for i in range(drug_length): 
    d = drugs[i]
    v = test.iloc[0][d]
    rowzerolist.append(v)

rowzerolist

for i in range(drug_length): 
    d = drugs[i]
    v = test.iloc[1][d]
    rowonelist.append(v)
    
rowonelist    

for i in range(drug_length): 
    v = round(((rowonelist[i] - rowzerolist[i])/rowzerolist[i]) * 100, 2)
    pctlist.append(v)
    
pctlist

chgframe = pd.DataFrame(list(zip(drugs, pctlist)), columns =['Drug', 'pct_chg']) 
chgframe
```

```python
# Store all Relevant Percent Changes into a Tuple
chgframe
chgframe2 = chgframe[chgframe["Drug"].isin(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'])] 
chgframe2


# Splice the data between passing and failing drugs
size_bins = [-100,0, 100]
group_names = ["Pass", "Fail"]
chgframe2["PassFail"] = pd.cut(x=chgframe2["pct_chg"], bins=size_bins, right=True, labels=group_names)
chgframe2.head()

xaxis = chgframe2["Drug"].values
yaxis = chgframe2["pct_chg"].values

chgframe2.loc[chgframe2.PassFail == "Pass", "color"] = "green"  
chgframe2.loc[chgframe2.PassFail == "Fail", "color"] = "red"
colors = chgframe2["color"].values

# Orient widths. Add labels, tick marks, etc. 
plt.bar(
    [1, 3,5,7],
    yaxis,
    align="center",
    width = 2,
    color=colors)

 
plt.xticks([2, 4,6,8],xaxis)

play =  [1, 3,5,7]

for i in range(len(xaxis)):
    if yaxis[i] > 0: 
        plt.text(play[i],5,str(int(round(yaxis[i],0))) + "%", horizontalalignment ='center', color="white", weight="bold")
    if yaxis[i] < 0: 
        plt.text(play[i],-5,str(int(round(yaxis[i],0))) + "%", horizontalalignment ='center', color="white", weight="bold")


        


plt.grid(True, which= "major")
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")


# Save the Figure
from io import BytesIO
f = "passfail.jpg"
plt.savefig(f, format="jpg")


plt.show()

# Use functions to label the percentages of changes


# Call functions to implement the function calls






```
